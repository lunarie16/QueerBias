```python
import matplotlib as plt
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.pyplot import figure
```

```python
import os 
os.chdir("../data/datasets")
```

```python
gender = pd.read_csv(
    'winoqueer_gender_identity.csv').set_index('Unnamed: 0')
sexuality = pd.read_csv('winoqueer_sexual_identity.csv').set_index('Unnamed: 0')
```

```python
gender.head()
```

```python
sexuality.head()
print(sexuality.Gender_ID_x.unique())
```

```python
blue = (0, 0.259, 0.510)
red = (0.918, 0.231, 0.025)
cyan = (0, 0.627, 0.666)
yellow = (1, 0.788, 0)
```

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i]+50, y[i], ha = 'center', fontsize = 'x-small')
figure(figsize=(8, 3))
value_counts = sexuality.Gender_ID_x.value_counts().sort_index(kind='mergesort').sort_values(ascending=False)
value_counts.plot.bar(color=blue, xlabel='', ylabel='Number of sentences')
addlabels(value_counts.index, value_counts.values)
plt.xticks(rotation=45)
plt.savefig("sexuality.pdf", format="pdf", bbox_inches="tight")
plt.show()
#
#
#




```

```python
figure(figsize=(8, 3))
value_counts2 = gender.Gender_ID_x.value_counts().sort_index(kind='mergesort').sort_values(ascending=False)
value_counts2.plot.bar(color=cyan, xlabel='', ylabel='Number of sentences')
addlabels(value_counts2.index, value_counts2.values)
plt.xticks(rotation=45)
plt.savefig("gender.pdf", format="pdf", bbox_inches="tight")

plt.show()
```

```python
winoqueer = pd.read_csv('winoqueer_final.csv')
val_counts = winoqueer.Gender_ID_x.value_counts().sort_values(ascending=False)
figure(figsize=(8, 3))
val_counts.plot.bar(color=red, xlabel='', ylabel='Number of sentences')
addlabels(val_counts.index, val_counts.values)
plt.xticks(rotation=45)
plt.savefig("original_winoqueer.pdf", format="pdf", bbox_inches="tight")
```
